```
class Statistician:
  def __init__(self):
    self.logic = 0
    self.memory = 0
    self.creativity = 0
    return self.initialization()

  def initialization(self):
    def get_sum():
      import random
      #max amount of 100 per attribute
      #choice of odd vals exclusing anything divisible by 5
      choices = [val for val in range(1,101,2) if val%5 != 0]
      choice1 = random.choice(choices)
      choice2 = random.choice(choices)
      choice3 = random.choice(choices)
      total = [choice1,choice2,choice3]
      #loop until sum matches desired range and return vals
      if sum(total) in range(111,178):
        self.logic = total[0]
        self.memory = total[1]
        self.creativity = total[2]
      else:
        get_sum()
    get_sum()
  

  #Funcitons from "Help the statisticians - part 2"

  #â€¯Square of a number: Create a function that returns the square of a number.
  def square(self, num):
    return num*num


  # Cube of a number: Create a function that returns the cube of a number.
  def cube(self, num):
    return num*num*num


  # Absolute value: Create a function that returns the absolute value of a number.
  def abs_val(self, num):
    if num < 0:
      return num*-1
    else:
      return num


  # Factorial of a number: Create a function that returns the factorial of a number.
  def factorial(self, num):
    # only positive numbers
    fact = 1
    for val in range(1,num+1):
      fact *= val
    return fact

  # Mode of a list of numbers: Create a function that returns the mode of a list of numbers.
  def mode(self, nums):
    nums = sorted(nums)
    if len(nums)%2 == 0:
      return (nums[int(len(nums)/2)]+nums[int((len(nums)/2)-1)])/2
    else:
      return nums[int(len(nums)/2)]


  # Average of a list of numbers: Create a function that returns the average of a list of numbers.
  def average(self, nums):
    len_nums = len(nums)
    sum = 0
    for vals in nums:
      sum += vals
    return sum/len_nums


  # Minimum of a list of numbers: Create a function that returns the minimum from within a list of numbers.
  def find_min(self, nums):
    import math
    min = math.inf 
    for val in nums:
      if val < min:
        min = val
    return min


  # Maximum of a list of numbers: Create a function that returns the maximum from within a list of numbers.
  def find_max(self, nums):
    import math
    max = -math.inf 
    for val in nums:
      if val > max:
        max = val
    return max


  # Power of x : Create a function that takes in two numbers as parameters, a number and a power, and returns that number to the indicated power.
  def num_to_power(self, num, pwr):
    n_p = 1
    for i in range(1,pwr+1):
      n_p *= num
    return n_p


  # Prime number less than x : Create a function that takes a number and returns all the prime numbers that are less than it.
  def find_primes(self, num):
    # num must be greater then 1
    if num < 2:
      return('num must be greater than 1')
    primes = set()
    for val1 in range(2,num+1):
      primes.add(val1)
      for val2 in range(2,val1):
        if (val1/val2) == (val1//val2):
          if val1 in primes:
            primes.remove(val1)
    return sorted(set(primes))


  #Continuation of Fibonnaci : Create a function that takes no parameters, but returns the first 30 numbers in the Fibonnaci sequence.
  def fibo(self, seq=30):
    #my parameter is just calling 30
    nums_in_fib = [0]
    a, b = 0, 1
    for i in range(seq-1):
      c = a + b
      nums_in_fib.append(c)
      a, b, c = c, a, b
    return nums_in_fib


  # Square root: Create a function that returns the square root of a number.
  def sq_root(self, num):
    return num**(1/2)
```

```
#Inherit from parent class
class dataAnalyst(Statistician):
  
  def __init__(self, name, first_name, age, gender, trainingPrevious):
    super().__init__()
    #class attributes
    self.name = str(name)
    self.first_name = str(first_name)
    self.age = int(age)
    self.gender = str(gender)
    self.trainingPrevious = str(trainingPrevious)
    self.motivation = int(100)
    self.progression = int(0)
  
  # class methods
  def train(self):
    #decreases motivation by a random value between 5 and 25 and increases progression by a random value between 20 and 30
    import random 
    mot_decrease = random.choice(range(5,26))
    prog_increase = random.choice(range(20,31))
    self.motivation = self.motivation - mot_decrease
    self.progression = self.progression + prog_increase

  def promotion(self):
    #increases motivation and progression both by a random value between 10 and 30
    import random 
    mot_increase = random.choice(range(10,31))
    prog_increase = random.choice(range(10,31))
    self.motivation = self.motivation + mot_increase
    self.progression = self.progression + prog_increase

  def fail(self):
    #decreases motivation by a random value between 20 and 40 and increases progression by a random value between 20 and 40
    import random 
    mot_decrease = random.choice(range(20,41))
    prog_increase = random.choice(range(20,41))
    self.motivation = self.motivation - mot_decrease
    self.progression = self.progression + prog_increase

  def success(self):
    #increases the motivation by a random value between 20 and 40 and increases the progression by a random value between 10 and 20
    import random 
    mot_increase = random.choice(range(20,41))
    prog_increase = random.choice(range(10,21))
    self.motivation = self.motivation + mot_increase
    self.progression = self.progression + prog_increase
  
  def play_game(self):
    """
    automatically and randomly launch methods and display a summarised status report for the attributes: motivation and progress
    program stops when the motivation falls to zero and displays "BRAVO YOU WON!!!" or when progress reaches 100 and then it displays "BRAVO YOU LEARNED!!!"
    """
    import random
    summary = []
    choices = [self.train, self.promotion, self.fail, self.success]

    while True:
      random.choice(choices)()
      summary.append((self.motivation, self.progression))
      #[print("motivation =", item[0], "progression =", item[1]) for item in summary]
      print("motivation =", self.motivation, "progression =", self.progression)

      if self.motivation < 0:
        print("BRAVO YOU WON!!!")
        break
      if self.progression >= 100:
        print("BRAVO YOU LEARNED!!!")
        break
```
